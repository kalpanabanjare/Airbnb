## Bivariate and Multivariate Analysis

```python
# Calculate the count of listings for each host
host_count = airbnb_df['host_id'].value_counts().head(10)

# plot a bar chart
plt.figure(figsize=(10, 6))
host_count.plot(kind='bar')
plt.xlabel('Host ID')
plt.ylabel('Number of Listings')
plt.title('Top 10 Hosts by Number of Listings')
plt.show()
```

```python
# price distribution by room type 
sns.barplot(data=airbnb_df, x='room_type', y='price')
plt.title('Price distribution by room type')
plt.show()
```

```python
# Create a stacked bar chart
plt.figure(figsize=(10, 6))
sns.countplot(data=airbnb_df, x='neighbourhood_group', hue='room_type')
plt.xlabel('Neighbourhood Group')
plt.ylabel('Count')
plt.title('Room Type Distribution by Neighbourhood Group')
plt.show()
```

```python
# Create bar plot for neighbourhood_group and price
plt.figure(figsize=(10,5))
sns.barplot(data=airbnb_df, x='neighbourhood_group', y='price')
plt.xlabel('Neighbourhood Group')
plt.ylabel('Price')
plt.title('Price Distribution by Neighbourhood Group')
plt.show()
```

```python
# Create box plot
plt.figure(figsize=(10,6))
sns.boxplot(data=airbnb_df, x='neighbourhood_group', y='price')
plt.xlabel('Neighbourhood Group')
plt.ylabel('Price')
plt.title('Price Distribution by Neighbourhood Group')
plt.show()
```

```python
# Calculate the average price for each neighbourhood group
avg_price_neighbourhood_group = airbnb_df.groupby('neighbourhood_group')['price'].mean().sort_values()

# Create bar plot
plt.figure(figsize=(10, 6))
avg_price_neighbourhood_group.plot(kind='bar')
plt.xlabel('Neighbourhood Group')
plt.ylabel('Average Price')
plt.title('Average Price by Neighbourhood Group')
plt.show()
```

```python
# ploting scatter 
plt.figure(figsize=(10, 5))
plt.scatter(airbnb_df['minimum_nights'], airbnb_df["number_of_reviews"])
plt.xlabel('minimum_nights')
plt.ylabel('number_of_reviews')
plt.title('number_of_reviews by minimun_nights')
plt.show()
```

```python
# Grouping data by neighborhood group and room type
# calculate their count
room_type_counts = airbnb_df.groupby(['neighbourhood_group', 'room_type']).size().unstack()

# Plotting bar chart
room_type_counts.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Distribution of Room Types according to Neighborhood Groups')
plt.xlabel('Neighborhood Group')
plt.ylabel('Count')
plt.legend(title='Room Type')
plt.show()
```

```python
# Calculate the availability by neighborhood_group
availability = airbnb_df.groupby('neighbourhood_group')['availability_365'].mean().sort_values()

# ploting 
plt.figure(figsize=(10, 6))
plt.bar(availability.index, availability)
plt.xlabel('Neighborhood Group')
plt.ylabel('Availability (Days)')
plt.title('Availability of Neighbourhood Group')
plt.show()
```

```python
# list the price over $500 
filtere_price_data = airbnb_df[airbnb_df['price'] <= 500]

# ploting histogram
plt.figure(figsize=(10, 6))
plt.hist(filtere_price_data['price'], bins=20)
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Preferred Price Range of Customers')
plt.show()
```

```python
# list the price over $500 
filtere_price_data = airbnb_df[airbnb_df['price'] <= 500]

# ploting box plot
plt.figure(figsize=(10, 5))
sns.boxplot(x='neighbourhood_group', y='price', hue='room_type', data=filtere_price_data)
plt.xlabel('Neighbourhood Group')
plt.ylabel('Price')
plt.title('Price variation based on Room_type and Neighbourhood_group')
plt.show()
```

```python
# Calculateing the number of bookings for each neighborhood
bookings_by_neighborhood = airbnb_df['neighbourhood'].value_counts()

# Select the top 5 neighborhoods based on bookings
top5_neighborhoods_bookings = bookings_by_neighborhood.head(5)

print(top5_neighborhoods_bookings)
```

```python
# Plotting the top neighborhoods based on bookings
plt.figure(figsize=(10, 5))
top5_neighborhoods_bookings.plot(kind='bar')
plt.xlabel('Neighborhood')
plt.ylabel('Number of Bookings')
plt.title('Top 5 Neighborhoods based on Bookings')
plt.show()
```

```python
# Calculateing number of reviews for each neighborhood
reviews_by_neighborhood = airbnb_df.groupby('neighbourhood')['number_of_reviews'].sum()

# Select the top 5 neighborhoods based on reviews
top5_neighborhoods_reviews = reviews_by_neighborhood.head(5)
print(top5_neighborhoods_reviews)
```

```python
# Plotting the top neighborhoods based on reviews
plt.figure(figsize=(10, 5))
top5_neighborhoods_reviews.plot(kind='bar')
plt.xlabel('Neighborhood')
plt.ylabel('Number of Reviews')
plt.title('Top 5 Neighborhoods based on Reviews')
plt.show()
```
